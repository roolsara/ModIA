## Pr√©liminaires

```python
# The function to learn
def f(x):
    return torch.abs(x) * torch.sin(2*2*torch.pi * x)
```

```python
P = 1000
x_espace = torch.linspace(-1,1,P)

plt.plot(x_espace, f(x_espace))
```

```python
N = 100
x = torch.linspace(-1,1,N)
y = f(x)
```

## Code principal

```python
class RandomReLU(nn.Module):
    def __init__(self, D = 10):
        super().__init__()
        self.b = 2 * (torch.rand(D)-0.5).to(device)
        self.weight = torch.nn.Parameter(torch.zeros(D))

    def forward(self, x):
        dif = x - self.b[None, :]
        hidden = torch.maximum(dif, torch.zeros_like(dif))
        
        return torch.sum(self.weight[None, :] * hidden, dim=1)
```

```python
model = RandomReLU(D=1000)
```

```python
def count_parameters(model):
    return sum(p.numel() for p in model.parameters() if p.requires_grad)

print("Number of parameters: ", count_parameters(model))
```

```python
#optimizer = torch.optim.SGD(model.parameters(), lr=0.001, momentum=0.9)
optimizer = optim.Adam(model.parameters(), lr=0.001, betas=(0.9, 0.999))

```

```python
from tqdm.auto import tqdm
```

```python
num_train_iteration = 10000
t = tqdm(range(num_train_iteration))
#avg_loss = 0

batch_size = 200
t = tqdm(range(num_train_iteration))
loss_tracking = []
for i in t:
    idx = torch.randint(0, len(x),(batch_size,))
    x_batch = x[idx, None]
    y_batch = y[idx]

    optimizer.zero_grad()
    y_pred = model(x_batch)
    loss = torch.mean((y_pred - y_batch)**2)
    #loss = torch.mean(torch.abs(y_pred - y_batch)) pas bien
    loss.backward()
    optimizer.step()
    loss_tracking.append(loss.item())

    if i% 1000 == 0:
        with torch.no_grad():
            fig, axs = plt.subplots(1, 2, figsize=(10,5))
            axs[0].plot(x_espace.cpu().numpy(), f(x_espace).cpu().numpy())
            axs[1].plot(x_espace.cpu().numpy(), model(x_espace[:,None]).cpu().numpy())
            plt.show()

plt.figure()
plt.plot(loss_tracking)
```
