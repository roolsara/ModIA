Dans ce TP, nous allons avoir besoin de la fonction `classify_by_LDA()` codée au TP 1. Vous pouvez la recopier dans la cellule suivante. 

```python

def classify_by_LDA(dataTrain: dict, dataTest: dict) -> dict:
    """
    Trains a Linear Discriminant Analysis (LDA) classifier using training data,
    evaluates it on both training and test sets, and returns performance metrics.

    Parameters:
        dataTrain (dict): Training data as a dictionary with class indices as keys
                          and feature arrays as values.
        dataTest (dict): Test data in the same format as dataTrain.

    Returns:
        dict: A dictionary containing:
            - 'train_cm': Confusion matrix on training data
            - 'train_acc': Accuracy on training data
            - 'test_cm': Confusion matrix on test data
            - 'test_acc': Accuracy on test data
            - 'lda': The trained LDA classifier object
    """
    ## convert class-wise dictionaries into flat feature and label arrays
    featuresTrain, labelsTrain = convert_cls2fls(dataTrain, NbClasses, NbMapsTrain)
    featuresTest, labelsTest = convert_cls2fls(dataTest, NbClasses, NbMapsTest)

    clf = LinearDiscriminantAnalysis()  ## create an instance of LDA classifier

    clf.fit(featuresTrain, labelsTrain)  ## train the classifier using training data

    ## make predictions on both training and test sets
    predsTrain = clf.predict(featuresTrain)
    predsTest = clf.predict(featuresTest)

    ## compute confusion matrices to assess classification results
    cmatTrain = confusion_matrix(labelsTrain, predsTrain)
    cmatTest = confusion_matrix(labelsTest, predsTest)

    ## compute accuracy scores for both training and test
    accuracyTrain = accuracy_score(labelsTrain, predsTrain)
    accuracyTest = accuracy_score(labelsTest, predsTest)

    ## return results in a dictionary 
    return {
        'train_cm': cmatTrain,
        'train_acc': accuracyTrain,
        'test_cm': cmatTest,
        'test_acc': accuracyTest,
        'lda': clf
    }
```

## 1. Calcul des coefficients de Scattering de 1er ordre

Nous allons utiliser la fonction `Scattering2D()` de la librairie Kymatio. Voir la doc https://www.kymat.io/codereference.html pour la définition des coefficients calculés. 

On définit la fonction suivante pour se restreindre aux coefficients d'ordres 0 et 1:

```python
scattering1 = Scattering2D(shape=(SizeMaps, SizeMaps), J=J, L=L, max_order=1)
if use_cuda:
    scattering1 = scattering1.cuda()  # On passe la fonction sur GPU
```

**1.1 TODO :** Compléter la fonction suivante qui calcule les coefficients de scattering d'ordre 1 sur un jeu d'images. Il faudra penser à convertir l'image numpy array en Tenseur torch en utilisant la fonction `torch.from_numpy()`. Afin de tenir compte de l'invariance par translation sur l'image, on moyennera les coefficients sur les blocs de pixels. Appliquer cette fonction aux images d'entraînement et de validation. 

```python
def compute_features_o1(Imgs: dict, NbClasses: int, NbMaps: int, J: int, L: int) -> dict:
    scato1_cls = {}
    num_features = 1 + (J * L)

    for i in range(1, NbClasses + 1): # Boucle sur les classes
        scato1 = np.zeros((NbMaps, num_features))
        for j in range(NbMaps): # Boucle sur les images
            img = Imgs[i][j] # Supposons que Imgs est un dictionnaire de listes d'images
            img_tensor = torch.from_numpy(img).float() # Conversion en tenseur PyTorch

            if use_cuda:
                img_tensor = img_tensor.cuda()

            scato = scattering1(img_tensor)
            scato1[j, 0] = scato[0].mean().cpu().numpy()

            if scato.shape[0] > 1 :
                order1_coeffs = scato[1:].mean(dim=(1,2))
                scato1[j, 1:1+order1_coeffs.numel()] = order1_coeffs.flatten().cpu().numpy()

        scato1_cls[i] = scato1
    
    return scato1_cls
```

# 2. Calcul des coefficients de scattering de 2nd ordre

On définit la fonction suivante pour inclure les ordres 0, 1 et 2:

```python
scattering2 = Scattering2D(shape=(SizeMaps, SizeMaps), J=J, L=L, max_order=2)
if use_cuda:
    scattering2 = scattering2.cuda()  # On passe la fonction sur GPU
```

**2.1 TODO :** Comme à la partie précédente, compléter la fonction suivante qui calcule les coefficients de scattering d'ordre 0, 1 et 2. Appliquer la fonction aux données d'entraînement et de test.

```python
def compute_features_o2(Imgs: dict, NbClasses: int, NbMaps: int, J: int, L: int, SizeMaps: int) -> dict:
    scato2_cls = {}
    num_features = 1 + J * L + (J * (J - 1) * L * L // 2) # Number of coefficients (orders 0, 1, 2)

    # Initialize the scattering transform
    scattering2 = Scattering2D(J=J, shape=(SizeMaps, SizeMaps), L=L, max_order=2)
    if use_cuda:
        scattering2 = scattering2.cuda()

    for i in range(1, NbClasses + 1):
        scato2 = np.zeros((NbMaps, num_features))
        for j in range(NbMaps):
            img = Imgs[i][j]
            img_tensor = torch.from_numpy(img).float()

            if use_cuda:
                img_tensor = img_tensor.cuda()


            scato = scattering2(img_tensor)

            scato2[j, 0] = scato[0].mean().cpu().numpy()

            if scato.shape[0] > 1:
                order1_coeffs = scato[1:J*L+1].mean(dim=(1, 2))
                scato2[j, 1:1 + order1_coeffs.numel()] = order1_coeffs.flatten().cpu().numpy()

            if scato.shape[0] > J*L + 1:
                order2_coeffs = scato[J*L+1:].mean(dim=(1, 2))
                scato2[j, 1 + J*L:1 + J*L + order2_coeffs.numel()] = order2_coeffs.flatten().cpu().numpy()

        scato2_cls[i] = scato2

    return scato2_cls
```
